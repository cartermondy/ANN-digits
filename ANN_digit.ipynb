# HMW 4 NN OCR Classification

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
```

```python
# Load dataset 
mnist = pd.read_csv('./MNIST_HW4.csv')
mnist.tail()
```

```python
# Prepare data (features and labels)
X = mnist.drop(columns='label').values
y = mnist['label'].values
```

```python
# Define neural network 
def create_nn():
    # Create model
    model = tf.keras.models.Sequential()
    # Input layer
    model.add(tf.keras.layers.Dense(128, activation='relu', input_shape=(784, )))
    #Hidden layers
    model.add(tf.keras.layers.Dense(128, activation='relu'))
    model.add(tf.keras.layers.Dense(128, activation='relu'))
    # Output layer (10 units for 10 digits that it could be)
    model.add(tf.keras.layers.Dense(10, activation='softmax'))
    # Compile model
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Perform 5-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Create and train neural network
    model = create_nn()
    model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))

    # Make predictions on the test set
    y_pred = np.argmax(model.predict(X_test), axis=1)

    # Calculate accuracy and append value to list
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
```

# Accuracies Across Folds

```python
# Accuracies for each fold
epoch_num = 1
for i in accuracies:
    print(f'Accuracy Score for Epoch {epoch_num}: {i}')
    epoch_num += 1
# Average accuracy
average_accuracy = np.mean(accuracies)
print(f'Average Accuracy: {average_accuracy}')
```
